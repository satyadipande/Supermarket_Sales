```python
# Now let's see that which gender purschased which product line 
pd.crosstab(supermarket_sales["Product line"], supermarket_sales["Gender"])
```

```python
# Create a plot
pd.crosstab(supermarket_sales["Product line"], supermarket_sales["Gender"]).plot(kind="bar",
                                    figsize=(10, 6),
                                    color=["salmon", "lightblue"])
                                                                                     
                                                                                    
# Communicate our work
plt.title("Analysis of preffered Product line for both Gender")
plt.xlabel("Gender")
plt.ylabel("Quantity")
plt.legend(["Female", "Male"]);
```

On the basis of above analysis it can be clearly seen that men purchased more of Health and beauty products whereas woman purchased more of Fashion accessories.

Also let's check at what time of the day are sales maximum and of which product category

```python
product_count = sns.relplot(x="Time", y="Quantity", col='Product line', row='Branch', estimator=None, kind="line", data=supermarket_sales);
```

From the above it can be seen that food and beverages were the most sold product line and sales went high after 7 pm.
